# Relational Algebra!

### Load data and display first 10 entries

```python
from reframe import Relation

fortune_500_data = Relation("fortune5002017.csv", sep = "|")

fortune_500_data.head(10)
```

### Project only company titles instead of entire row

```python
fortune_500_data.project(["title"]).head(20)
```

### Project several columns 

```python
fortune_500_data.project(["rank", "title", "profits", "employees"]).head(20)
```

## Filter and display companies with less than 3000 employees

```python
fortune_500_data.query("employees < 3000")
```

### Combine operators into a pipeline

1) Project columns rank, title, profits, employees, and industry

2) Query on profits greater than 10,000

3) Sort by industry first and number of employees second (companies in the same industry will be sorted by the number of employees they have)

```python
fortune_500_data.project(["rank", "title", "profits", "employees", "industry"])\
    .query("profits > 10000")\
    .sort(["industry", "employees"], ascending = False)
```

### Another pipeline example

1) Project columns title, rank, previousrank, and revenues

2) Query on rank less than previousrank and revenues less than 7000

3) Sort by rank

```python
fortune_500_data.project(["title", "rank", "previousrank", "revenues"])\
    .query("rank < previousrank and revenues < 7000")\
    .sort(["rank"], ascending = True)
```

### Groupby Example

1) Combine all rows with the same industry into one row

2) Use mean aggregate function to find the mean revenue for each individual industry

3) Sorty by mean_revenues

```python
fortune_500_data.groupby(['industry']).mean('revenues')\
     .sort(["mean_revenues"], ascending = False)
```
